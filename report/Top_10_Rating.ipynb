```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.ion()
matplotlib.is_interactive()
```

```python
data = pd.read_csv("./data/top_10_rating.csv")
```

```python
data.head()
```

```python
Y_axis = np.arange(len(data["name"]))

plt.barh(Y_axis + 0.2, data["rating_average"], 0.4)

plt.yticks(Y_axis, data["name"])
plt.xlabel("Rating Average")
plt.ylabel("Products")
plt.title("Top 10 Rating")
plt.ticklabel_format(style='plain', axis='x')
plt.show()
```
