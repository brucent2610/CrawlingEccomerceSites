```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib as mpl

plt.ion()
matplotlib.is_interactive()
mpl.rcParams['figure.figsize'] = (20, 10)
mpl.rcParams['font.size'] = 11.0
```

```python
data = pd.read_csv("./data/categories_count.csv")
```

```python
data_filtered = data.head(5)
```

```python
Y_axis = np.arange(len(data_filtered["_id"]))

plt.barh(Y_axis + 0.2, data_filtered["count"], 0.4)

plt.yticks(Y_axis, data_filtered["_id"])
plt.xlabel("Categories")
plt.ylabel("Products")
plt.title("Categories Products")
plt.ticklabel_format(style='plain', axis='x')
plt.show()
```

```python
data['percent'] = (data['count'] / data['count'].sum()) * 100
```

```python
data_pied = data.head()
```

```python
plt.pie(data_pied['percent'], labels=data_pied['_id'] , autopct='%1.1f%%', startangle=90)

# Set aspect ratio to be equal so that pie is drawn as a circle
plt.axis('equal')

# Set title
plt.title('Pie Chart')

# Show the plot
plt.show()
```
